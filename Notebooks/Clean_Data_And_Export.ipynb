```python

# Ensure date columns are in datetime format
df['Order_Date'] = pd.to_datetime(df['Order_Date'])
df['Ship_Date'] = pd.to_datetime(df['Ship_Date'])

# --- Step 1: Create Dimension Tables ---

# 1. Sub_Category
sub_category_df = df[['Sub_Category']].drop_duplicates().reset_index(drop=True)
sub_category_df['sub_category_id'] = sub_category_df.index + 1
sub_category_df = sub_category_df.rename(columns={'Sub_Category': 'sub_category'})

# 2. Category
category_df_temp = df[['Category', 'Sub_Category']].drop_duplicates().reset_index(drop=True)
category_df_temp = pd.merge(category_df_temp, sub_category_df, left_on='Sub_Category', right_on='sub_category', how='left')
category_df = category_df_temp.rename(columns={'Category': 'category'})
category_df['category_id'] = category_df.index + 1
category_df = category_df[['category_id', 'category', 'sub_category_id']]

# 3. Product (and a new junction table for many-to-many relationship)
product_df = df[['Product_ID', 'Product_Name']].drop_duplicates().reset_index(drop=True)
product_df = product_df.rename(columns={'Product_ID': 'product_id', 'Product_Name': 'product_name'})
# Create the junction table to handle product-category many-to-many relationship
product_to_category_df = df[['Product_ID', 'Category']].drop_duplicates().reset_index(drop=True)
product_to_category_df = pd.merge(product_to_category_df, category_df[['category', 'category_id']], left_on='Category', right_on='category', how='left')
product_to_category_df = product_to_category_df[['Product_ID', 'category_id']]
product_to_category_df = product_to_category_df.rename(columns={'Product_ID': 'product_id'})

# 4. Segment
segment_df = df[['Segment']].drop_duplicates().reset_index(drop=True)
segment_df = segment_df.rename(columns={'Segment': 'segment'})
segment_df['segment_id'] = segment_df.index + 1

# 5. Region
region_df = df[['Region']].drop_duplicates().reset_index(drop=True)
region_df = region_df.rename(columns={'Region': 'region'})
region_df['region_id'] = region_df.index + 1

# 6. State
state_df = df[['State', 'Region']].drop_duplicates().reset_index(drop=True)
state_df = pd.merge(state_df, region_df, left_on='Region', right_on='region', how='left')
state_df = state_df.rename(columns={'State': 'state'})
state_df['state_id'] = state_df.index + 1
state_df = state_df[['state_id', 'state', 'region_id']]

# 7. City
city_df = df[['City', 'State']].drop_duplicates().reset_index(drop=True)
city_df = pd.merge(city_df, state_df[['state', 'state_id']], left_on='State', right_on='state', how='left')
city_df = city_df.rename(columns={'City': 'city'})
city_df['city_id'] = city_df.index + 1
city_df = city_df[['city_id', 'city', 'state_id']]

# 8. Postal_Code
postal_code_df = df[['Postal_Code', 'City']].drop_duplicates().reset_index(drop=True)
postal_code_df = pd.merge(postal_code_df, city_df[['city', 'city_id']], left_on='City', right_on='city', how='left')
postal_code_df = postal_code_df.rename(columns={'Postal_Code': 'postal_code'})
postal_code_df['postal_code_id'] = postal_code_df.index + 1
postal_code_df = postal_code_df[['postal_code_id', 'postal_code', 'city_id']]

# 9. Country
country_df = df[['Country', 'Region']].drop_duplicates().reset_index(drop=True)
country_df = pd.merge(country_df, region_df[['region', 'region_id']], left_on='Region', right_on='region', how='left')
country_df = country_df.rename(columns={'Country': 'country'})
country_df['country_id'] = country_df.index + 1
country_df = country_df[['country_id', 'country', 'region_id']]

# 10. Customer
customer_df = df[['Customer_ID', 'Customer_Name', 'Segment', 'Country']].drop_duplicates(subset=['Customer_ID']).reset_index(drop=True)
customer_df = pd.merge(customer_df, segment_df, left_on='Segment', right_on='segment', how='left')
customer_df = pd.merge(customer_df, country_df[['country', 'country_id']], left_on='Country', right_on='country', how='left')
customer_df = customer_df[['Customer_ID', 'Customer_Name', 'segment_id', 'country_id']]
customer_df = customer_df.rename(columns={'Customer_ID': 'customer_id', 'Customer_Name': 'customer_name'})

# --- Step 2: Create Fact and Junction Tables ---

# 11. Orders Fact Table
orders_df = df[['Order_ID', 'Sales', 'Quantity', 'Discount', 'Profit', 'Customer_ID', 'Order_Date', 'Ship_Mode', 'Ship_Date']].drop_duplicates(subset=['Order_ID']).reset_index(drop=True)
orders_df = orders_df.rename(columns={'Order_ID': 'order_id', 'Customer_ID': 'customer_id', 'Order_Date': 'order_date', 'Ship_Date': 'ship_date'})

# 12. Products_to_orders Junction Table
products_to_orders_df = df[['Order_ID', 'Product_ID']].drop_duplicates().reset_index(drop=True)
products_to_orders_df = products_to_orders_df.rename(columns={'Order_ID': 'order_id', 'Product_ID': 'product_id'})

# --- Step 3: Export all DataFrames to CSVs ---
output_dir = os.path.join("..", "Data", "Clean")
os.makedirs(output_dir, exist_ok=True)

# Exporting the DataFrames
orders_df.to_csv(os.path.join(output_dir, 'orders.csv'), index=False)
customer_df.to_csv(os.path.join(output_dir, 'customer.csv'), index=False)
product_df.to_csv(os.path.join(output_dir, 'product.csv'), index=False)
country_df.to_csv(os.path.join(output_dir, 'country.csv'), index=False)
category_df.to_csv(os.path.join(output_dir, 'category.csv'), index=False)
region_df.to_csv(os.path.join(output_dir, 'region.csv'), index=False)
state_df.to_csv(os.path.join(output_dir, 'state.csv'), index=False)
city_df.to_csv(os.path.join(output_dir, 'city.csv'), index=False)
postal_code_df.to_csv(os.path.join(output_dir, 'postal_code.csv'), index=False)
sub_category_df.to_csv(os.path.join(output_dir, 'sub_category.csv'), index=False)
segment_df.to_csv(os.path.join(output_dir, 'segment.csv'), index=False)
products_to_orders_df.to_csv(os.path.join(output_dir, 'Products_to_orders.csv'), index=False)
product_to_category_df.to_csv(os.path.join(output_dir, 'product_to_category.csv'), index=False)

print(f"All DataFrames exported to CSVs in '{output_dir}'")
```

```python

# Ensure date columns are in datetime format
df['Order_Date'] = pd.to_datetime(df['Order_Date'])
df['Ship_Date'] = pd.to_datetime(df['Ship_Date'])

# --- Step 1: Create Dimension Tables ---

# 1. Location Dimension Table
location_df = df[['Postal_Code', 'City', 'State', 'Country', 'Region']].drop_duplicates().reset_index(drop=True)
location_df = location_df.rename(columns={'Postal_Code': 'postal_code', 'City': 'city', 'State': 'state', 'Country': 'country', 'Region': 'region'})
location_df['location_id'] = location_df.index + 1

# 2. Sub_Category
sub_category_df = df[['Sub_Category']].drop_duplicates().reset_index(drop=True)
sub_category_df = sub_category_df.rename(columns={'Sub_Category': 'sub_category'})
sub_category_df['sub_category_id'] = sub_category_df.index + 1

# 3. Category
category_df_temp = df[['Category', 'Sub_Category']].drop_duplicates().reset_index(drop=True)
category_df = pd.merge(category_df_temp, sub_category_df, left_on='Sub_Category', right_on='sub_category', how='left')
category_df = category_df.rename(columns={'Category': 'category'})
category_df['category_id'] = category_df.index + 1
category_df = category_df[['category_id', 'category', 'sub_category_id']]

# 4. Product
product_df = df[['Product_ID', 'Product_Name']].drop_duplicates(subset=['Product_ID']).reset_index(drop=True)
product_df = product_df.rename(columns={'Product_ID': 'product_id', 'Product_Name': 'product_name'})

# 5. Junction table for Product to Category
product_to_category_df = df[['Product_ID', 'Category']].drop_duplicates().reset_index(drop=True)
product_to_category_df = pd.merge(product_to_category_df, category_df[['category', 'category_id']], left_on='Category', right_on='category', how='left')
product_to_category_df = product_to_category_df[['Product_ID', 'category_id']]
product_to_category_df = product_to_category_df.rename(columns={'Product_ID': 'product_id'})

# 6. Segment
segment_df = df[['Segment']].drop_duplicates().reset_index(drop=True)
segment_df = segment_df.rename(columns={'Segment': 'segment'})
segment_df['segment_id'] = segment_df.index + 1

# 7. Customer
customer_df = df[['Customer_ID', 'Customer_Name', 'Segment']].drop_duplicates(subset=['Customer_ID']).reset_index(drop=True)
customer_df = pd.merge(customer_df, segment_df, left_on='Segment', right_on='segment', how='left')
customer_df = customer_df[['Customer_ID', 'Customer_Name', 'segment_id']]
customer_df = customer_df.rename(columns={'Customer_ID': 'customer_id', 'Customer_Name': 'customer_name'})

# --- Step 2: Create Fact Table ---

# 8. Orders Fact Table
orders_df = df[['Order_ID', 'Sales', 'Quantity', 'Discount', 'Profit', 'Customer_ID', 'Order_Date', 'Ship_Mode', 'Ship_Date', 'Postal_Code', 'City', 'State', 'Country', 'Region']].drop_duplicates(subset=['Order_ID']).reset_index(drop=True)

# Rename columns to snake_case before merging
orders_df = orders_df.rename(columns={
    'Order_ID': 'order_id',
    'Sales': 'sales',
    'Quantity': 'quantity',
    'Discount': 'discount',
    'Profit': 'profit',
    'Customer_ID': 'customer_id',
    'Order_Date': 'order_date',
    'Ship_Mode': 'ship_mode',
    'Ship_Date': 'ship_date',
    'Postal_Code': 'postal_code',
    'City': 'city',
    'State': 'state',
    'Country': 'country',
    'Region': 'region'
})

# Merge with location_df to add the location_id
orders_df = pd.merge(orders_df, location_df, on=['postal_code', 'city', 'state', 'country', 'region'], how='left')

# Select the final columns
orders_df = orders_df[['order_id', 'sales', 'quantity', 'discount', 'profit', 'customer_id', 'order_date', 'ship_mode', 'ship_date', 'location_id']]

# 9. Products_to_orders Junction Table
products_to_orders_df = df[['Order_ID', 'Product_ID']].drop_duplicates().reset_index(drop=True)
products_to_orders_df = products_to_orders_df.rename(columns={'Order_ID': 'order_id', 'Product_ID': 'product_id'})

# --- Step 3: Export all DataFrames to CSVs ---
output_dir = os.path.join("..", "Data", "Clean")
os.makedirs(output_dir, exist_ok=True)

orders_df.to_csv(os.path.join(output_dir, 'orders.csv'), index=False)
customer_df.to_csv(os.path.join(output_dir, 'customer.csv'), index=False)
product_df.to_csv(os.path.join(output_dir, 'product.csv'), index=False)
location_df.to_csv(os.path.join(output_dir, 'location.csv'), index=False)
category_df.to_csv(os.path.join(output_dir, 'category.csv'), index=False)
sub_category_df.to_csv(os.path.join(output_dir, 'sub_category.csv'), index=False)
segment_df.to_csv(os.path.join(output_dir, 'segment.csv'), index=False)
products_to_orders_df.to_csv(os.path.join(output_dir, 'products_to_orders.csv'), index=False)
product_to_category_df.to_csv(os.path.join(output_dir, 'product_to_category.csv'), index=False)

print(f"All DataFrames exported to CSVs in '{output_dir}'")
```
